#### Clase Auto

```python
class Auto:
    def __init__(self, color, peso, tamanio, alto, largo, n_ruedas, n_puertas, tipo):
        self.color = color
        self.peso = peso
        self.tamanio = tamanio
        self.alto = alto
        self.largo = largo
        self.n_ruedas = n_ruedas
        self.n_puertas = n_puertas
        self.tipo = tipo
        self.estado = Estado()
        self.accion = Accion(self.estado)

    def get_auto(self):
        return {
            "color": self.color,
            "peso": self.peso,
            "tamanio": self.tamanio,
            "alto": self.alto,
            "largo": self.largo,
            "n_ruedas": self.n_ruedas,
            "n_puertas": self.n_puertas,
            "tipo": self.tipo
        }
```

#### Clase Estado

```python
class Estado:
    def __init__(self, arrancado=False, detenido=True, circulando=False, estacionado=False, daniado=False):
        self.arrancado = arrancado
        self.detenido = detenido
        self.circulando = circulando
        self.estacionado = estacionado
        self.daniado = daniado

    def get_estado(self):
        return vars(self)
```

#### Clase Acción

```python
class Accion:
    def __init__(self, estado):
        self.estado = estado

    @property
    def estado(self):
        return self._estado

    @estado.setter
    def estado(self, nuevo_estado):
        self._estado = nuevo_estado

    def accion_arrancar(self):
        self.estado.arrancado = True
        self.estado.detenido = False
        self.estado.circulando = False
        self.estado.estacionado = False

    def accion_frenar(self):
        self.estado.circulando = False
        self.estado.detenido = True
        self.estado.estacionado = False

    def accion_acelerar(self):
        self.estado.detenido = False
        self.estado.circulando = True
        self.estado.estacionado = False

    def accion_estacionar(self):
        self.estado.detenido = True
        self.estado.circulando = False
        self.estado.estacionado = True
```
