```python
def procesa_lista_notas(notas):
    aprueban = [nota for nota in notas if nota >= 60]
    print(f"Promedio de todos los estudiantes: {round(sum(notas) / len(notas),1)}")
    print(f"Estudiantes que aprobaron: {len(aprueban)}")
    print(f"Promedio de los estudiantes que aprobaron: {round(sum(aprueban) / len(aprueban),1)}")
```

```python
procesa_lista_notas([33, 35, 44, 36, 88, 99, 87, 60])
```
