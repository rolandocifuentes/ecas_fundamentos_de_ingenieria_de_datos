### Manejo de valores nulos y duplicados

**Identificaci贸n y conteo de valores nulos**

```python
df.isnull().sum()
```

**Identificaci贸n y conteo de valores NA**

```python
df.isna().sum()
```

**Imputa los valores faltantes en columnas num茅ricas con la media y en columnas categ贸ricas con la moda**

```python
df.fillna({"Ronda":df["Ronda"].mean(),
           "Goles Local":df["Goles Local"].mean(),
           "Goles Visitante":df["Goles Visitante"].mean(),
           "Total Goles":df["Total Goles"].mean(),
           "Fecha":df["Fecha"].mode()[0],
           "Local":df["Local"].mode()[0],
           "Vistante":df["Vistante"].mode()[0],
           "Resultado":df["Resultado"].mode()[0],
          },
          inplace=True)          
```

**Identificaci贸n y conteo de registros duplicados**

```python
df.duplicated().sum()
```

```python
# ver duplicados
df[df.duplicated()]
```

```python
# Eliminar registros duplicados
df.drop_duplicates(inplace=True)
```

### Creaci贸n de nuevas columnas y transformaci贸n de datos

**Crea etiqueta Goles Visitante**

```python
df["Goles Diferencia"] = df["Goles Local"] - df["Goles Visitante"]
df.head()
```

```python
# Ordena Etiquetas 
# Obtener todas las columnas
cols = df.columns.tolist()
# Encontrar la posici贸n de 'Resultado'
idx = cols.index("Resultado")
```

```python
# Insertar 'Goles Diferencia' justo antes
cols.insert(idx, cols.pop(cols.index("Goles Diferencia")))
# Reordenar el DataFrame
df = df[cols]
df.head()
```

<br>

**Convierte el tipo de datos de una columna**

```python
# Divide y convierte la fecha
df["Fecha"] = pd.to_datetime(df["Fecha"])
df["A帽o"] = df["Fecha"].dt.year
df["Mes"] = df["Fecha"].dt.month
df["D铆a"] = df["Fecha"].dt.day
df.head()
```

```python
# ordena etiquetas DataFrame
cols = df.columns.tolist()
```

```python
cols.insert(2, cols.pop(cols.index("A帽o")))
cols.insert(3, cols.pop(cols.index("Mes")))
cols.insert(4, cols.pop(cols.index("D铆a")))
df = df[cols]
df.head()
```
